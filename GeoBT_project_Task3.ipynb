How does global alignment work of protein sequences work in Python?

```python
# install packages
from Bio.Seq import Seq
from Bio import SeqIO
from Bio import pairwise2
from Bio.SubsMat import MatrixInfo as matlist
from Bio.pairwise2 import format_alignment
from Bio import AlignIO
```

### read in fasta files

```python
#read in fasta files
sequences_fasta = SeqIO.parse("RdhA_Dehalococcoides_strain_CBDB1_sequence.fasta", "fasta")

#create empty lists
sequence_32RdhA = [] #sequences
sequence_32RdhA_id = [] #ids of the sequences


#save the sequences and their ids to 2 different lists
for sequence in sequences_fasta:
    sequence_32RdhA.append(str(sequence.seq)) #.seq returns the sequence
    sequence_32RdhA_id.append(str(sequence.id)) #.id returns the id of the sequence

# print two lists
print(sequence_32RdhA)
print(sequence_32RdhA_id)
```

### start of user menu

```python
#print start of user interface
print("Welcome to your global sequence alignment tool for RdhA in CBDB1")
print("\n")
print("The sequences are already loaded")
print("\n")
print("The global sequence alignment is executed with PAM30")
print("\n")

#store user input in variable
ans = int(input("Do you want to align all the sequences (1) or a pair of sequences (2)?: "))
```

### global alignment of sequences

```python
#load matrix and store in variable
matrix = matlist.pam30
#create empty list
score_list = []

#option comparison of all sequences with each other
if ans == 1: 
    #loop through sequence list to compare sequences with each other
    for i in range(len(sequence_32RdhA)-1):
        seq1 = sequence_32RdhA[i] #chose sequence 1 for aligning
        scoreseq1=[] #create empty list to store scores for one time looping through
        for j in range(1,len(sequence_32RdhA)):
            seq2 = sequence_32RdhA[j] #chose sequence 2 for aligning
            #aligning seq1 and seq2
            alignment1 = pairwise2.align.globaldx(seq1, seq2, matrix)[0] 
            scoreseq1.append(alignment1.score) #add score to one time loop score list
        #add one time looping score list to overall score list
        score_list.extend(scoreseq1)
    
    #print calculated store list
    print(score_list)
    #end of process   
    print("Done")
    print("\n")
    
# pairwise comparison of specific sequences
elif ans == 2: 

    while True:
        #store user input in variable for sequence 1 and 2
        seq1 = int(input("Please choose sequence no. 1 (1-32): "))
        seq2 = int(input("Please choose sequence no. 2 (1-32): "))

        #option input not recognized
        if seq1 not in range(1,33) or seq2 not in range(1,33):
            print("Number out of range, please chose no. between 1 and 32.")

        else:
           
            #aligning seq1 and seq2
            for a in pairwise2.align.globaldx(sequence_32RdhA[int(seq1)-1], sequence_32RdhA[int(seq2)-1], matrix):
                scoru = a.score #save score in variable
            #save alignments in variable
            alignments = pairwise2.align.globaldx(sequence_32RdhA[int(seq1)-1], sequence_32RdhA[int(seq2)-1], matrix)
            
            #set layout format to see alignment better
            m = 100 #places in one line
            for alignment in alignments:
                a,b,c,score,_ = pairwise2.format_alignment(*alignment).split("\n")
                #print alignment results
                for i in range(0,len(a),m):
                    print(a[i:i+m])
                    print(b[i:i+m])
                    print(c[i:i+m])
                    print()
                print(score) #print result
                print()
            
            #add score to score list
            score_list.append(scoru)
        
            #end of process
            print("Done")
            print("\n")
           
            break
```

### write results into new file

```python
if ans == 1: 
    while True:
        #save user input in variable
        y = input("Do you want to save your score (y/n)?: ").lower()

        #save results
        if y == "y":
            #Save scores into one file
            with open("user_score.txt", 'w') as fp:
                for item in score_list:
                    # write each item on a new line
                    fp.write("%s\n" % item)
                print('Done') #end of process
                print("Goodbye!")
            break
        #exit program
        elif y == "n":
            print("Goodbye!")
            break
        else: 
            print("Did not recognize your answer. Please try again")
elif ans ==2:
    while True:
        #save user input in variable
        y = input("Do you want to save your score (y/n)?: ").lower()

        #save results
        if y == "y":
            #define places in one line
            m=100
            #create and open file 
            with open('user_alignment.txt', 'w') as f:
                for alignment in alignments:
                    a,b,c,score,_ = pairwise2.format_alignment(*alignment).split("\n")
                    #write aligned sequences into file
                    for i in range(len(a)//m+1):
                        f.write(a[i:i+m]+'\n')
                        f.write(b[i:i+m]+'\n')
                        f.write(c[i:i+m]+'\n')
                        f.write('\n')
                    f.write(score+'\n') #write score into file
                    f.write('ids= '+sequence_32RdhA_id[seq1-1]+' and '+sequence_32RdhA_id[seq2-1]+'\n') #write seq id into file 
                    f.write('\n'+'\n')
                print('Done') #end of process
                print("Goodbye!")
            break
        #exit program
        elif y == "n":
            print("Goodbye!")
            break
        #option input not recognized
        else: 
            print("Did not recognize your answer. Please try again")
```
